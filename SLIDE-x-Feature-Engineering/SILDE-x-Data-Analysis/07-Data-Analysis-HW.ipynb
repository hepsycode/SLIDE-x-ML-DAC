```python
# import modules 
import matplotlib.pyplot as mp 
import pandas as pd 
import seaborn as sns 
import numpy as np
from pylab import savefig
```

```python
# import file with data 
data=pd.read_csv("Dataset/csv/Train/BambuTrainTest-Zynq-7000.csv") 
saveFolderImg = "Heatmap-Zynq-7000/"
```

```python
# DSP
DSP_code = data.iloc[:, np.r_[6, 14:48]] # Code Features
DSP_array = data.iloc[:, np.r_[6, 52:83]] # Code Features
DSP_matrix = data.iloc[:, np.r_[6, 84:115]] # Code Features
# print(DSP_matrix)
```

```python
# Calculate the correlation matrix
correlation_matrix = DSP_code.corr()

# Create the heatmap
mp.figure(figsize = (10,8))

# plotting correlation heatmap 
sns_plot = sns.heatmap(correlation_matrix, cmap='coolwarm', linecolor='white', linewidths=1) 

sns_plot.set_xticklabels(sns_plot.get_xmajorticklabels(), fontsize = 9)
sns_plot.set_yticklabels(sns_plot.get_ymajorticklabels(), fontsize = 9)

# sns.set(font_scale=1)
sns_plot.figure.savefig(saveFolderImg+'DSP_code.png', dpi=300, bbox_inches='tight')

correlation_matrix.to_csv(saveFolderImg+'DSP_code.csv')
  
# displaying heatmap 
mp.show()
```
