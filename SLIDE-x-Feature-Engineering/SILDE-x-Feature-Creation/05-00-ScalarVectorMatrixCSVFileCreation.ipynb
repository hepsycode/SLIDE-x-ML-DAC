```python
import time
import pandas as pd
from IPython.core.magic import register_cell_magic
from IPython.display import display, HTML

# CSV file to save execution times
csv_filename = path_ref_name + "05-00-ExecutionTime.csv"

# Initialize the CSV file with header if it doesn't exist
try:
    df = pd.read_csv(csv_filename)
except FileNotFoundError:
    df = pd.DataFrame(columns=["cell_number", "execution_time"])
    df.to_csv(csv_filename, index=False)

# Function to log execution times
def log_execution_time(cell_number, execution_time):
    df = pd.read_csv(csv_filename)
    new_row = pd.DataFrame({"cell_number": [cell_number], "execution_time": [execution_time]})
    df = pd.concat([df, new_row], ignore_index=True)
    df.to_csv(csv_filename, index=False)

# Decorator for the magic function
@register_cell_magic
def log_time(line, cell):
    start_time = time.time()
    exec(cell, globals())
    end_time = time.time()
    execution_time = end_time - start_time

    # Current cell number
    cell_number = int(line) if line.isdigit() else None
    if cell_number is None:
        display(HTML("<b>Warning:</b> Cell number not provided!"))
        return

    # Log the execution time
    log_execution_time(cell_number, execution_time)
    display(HTML(f"<b>Cell {cell_number}:</b> Execution time = {execution_time:.6f} seconds"))

# To use the magic function in the notebook
# %%log_time 1
# cell code
```

```python
%%log_time 4

import re
import os

# Function to classify the type of string
def classify_strings(lsubNP):
    variables = []
    indices = set()
    arrays = []
    matrices = []
    
    # First pass: classify variables and identify potential indices
    for s in lsubNP:
        if re.match(r'^[a-zA-Z_]\w*$', s):
            variables.append(s)
    
    # Second pass: classify arrays and matrices, and confirm indices
    for s in lsubNP:
        array_match = re.match(r'^([a-zA-Z_]\w*)\[(\w+)\]$', s)
        matrix_match = re.match(r'^([a-zA-Z_]\w*)\[(\w+)\]\[(\w+)\]$', s)
        if array_match:
            arrays.append(s)
            indices.add(array_match.group(2))
        elif matrix_match:
            matrices.append(s)
            indices.add(matrix_match.group(2))
            indices.add(matrix_match.group(3))
    
    # Remove indices from variables list
    variables = [v for v in variables if v not in indices]
    
    return variables, list(indices), arrays, matrices

data_list = []

dfInpTemp = pd.DataFrame()
dfInpTot = pd.DataFrame()

dfInpTempScalar = pd.DataFrame()
dfInpTempIndex = pd.DataFrame()
dfInpTempVector = pd.DataFrame()
dfInpTempMatrix = pd.DataFrame()

for ii in range(0, len(pathFile)):
    pF = pathFile[ii]
    print('pF:'+str(pF))

    with open(pF, 'r') as jf:
        data = jf.read()
    obj = json.loads(data)

    if os.name == 'nt':
        spR = rootFile[ii].split('\\')
        print('spR:'+str(spR))
    else:
        spR = rootFile[ii].split('/')
        print('spR:'+str(spR))
    
    function = spR[-1]
    print('function:'+str(function))
    # board = listD[i]
    # print('board:'+str(board))
    
        ################################# 
    #Device loop
    deviceList = list(obj.keys())
    print('deviceList:'+str(deviceList))
    deviceListR = []
    for u in range(0, len(deviceList)):
        if deviceList[u] in listD:
            deviceListR.append(deviceList[u])
            print('deviceListR:'+str(deviceList[u]))

    for n in range(0, len(deviceListR)):
        # Data_type loop
        device = deviceListR[n]
        print('device:'+str(device))
        dtarr = list(obj[deviceListR[n]].keys())
        print('dtarr:'+str(dtarr))
        
        dtarrR = []
        for u in range(0, len(dtarr)):
            if dtarr[u] in listDT:
                dtarrR.append(dtarr[u])
                print(f'dtarr[{u}]:'+str(dtarr[u]))
        
        for j in range(0, len(dtarrR)):
            dataype = dtarrR[j]
            print('dataype:'+str(dtarrR[j]))
            lsubNP = (list(obj[deviceListR[n]][dtarrR[j]].keys()))
            print('lsubNP:'+str(lsubNP))
            lsubVP = (list(obj[deviceListR[n]][dtarrR[j]].values()))
            print('lsubVP:'+str(lsubVP))
    
            for i in range(0, len(pathFileInp)):
                path = pathFileInp[i]

                if os.name == 'nt':
                    spRNN = rootFileInp[i].split('\\')
                else:
                    spRNN = rootFileInp[i].split('/')

                matrixVectorSizeNN = spRNN[-1] 
                optFlagNN = spRNN[-2]
                datatypeNN = spRNN[-3]
                deviceNN = spRNN[-4]
                # board = 'xc7a100t-1csg324-Artix-7-200-MHz'  # Bambu (HW)
                boardNN = spRNN[-4] # SW
                functionNN = spRNN[-5]

                if str(deviceNN) == str(device) and str(functionNN) == str(function) and str(dtarrR[j]) == str(datatypeNN) and str(optFlagNN) == 'optO1-00':
                    
                    try:
                        dfInpTemp = pd.read_csv(path, sep=',',engine='python')
                    except:
                        print(path)
                    
                    dfInpTemp['DEVICE'] = deviceNN
                    dfInpTemp['BOARD'] = boardNN
                    dfInpTemp['FUNCTION'] = functionNN
                    dfInpTemp['DATA_TYPE'] = datatypeNN
                    dfInpTemp['OPT_FLAG'] = optFlagNN
                    dfInpTemp['MV_SIZE'] = matrixVectorSizeNN
                    dfInpTemp['inputName'] = dfInpTemp['inputName'].map(lambda x: x.lstrip('values_'))
                    dfInpTemp['inputName'] = dfInpTemp['inputName'].astype(int)
                    dfInpTemp.sort_values('inputName')
            
                    # sublist dimension
                    ltot = len(lsubNP)
                    print('Folder: '+str(path))
                    
                    scalar_cols = []
                    index_cols = []
                    array_cols = []
                    matrix_cols = []
                    
                    variables, indices, arrays, matrices = classify_strings(lsubNP)
                    directory_path = os.path.dirname(path)
                    new_columns  = ['inputName']
                        
                    if variables:
                        all_columns = new_columns + variables
                        dfInpTemp[all_columns].to_csv(os.path.join(directory_path, 'ScalarVariable.csv'), index=False, encoding='utf-8')
                    if indices:
                        all_columns = new_columns + indices
                        dfInpTemp[all_columns].to_csv(os.path.join(directory_path, 'ScalarSize.csv'), index=False, encoding='utf-8')
                    if arrays:               
                        all_columns = new_columns + arrays
                        dfInpTemp[all_columns].to_csv(os.path.join(directory_path, 'Vector.csv'), index=False, encoding='utf-8')
                    if matrices:
                        all_columns = new_columns + matrices
                        dfInpTemp[all_columns].to_csv(os.path.join(directory_path, 'Matrix.csv'), index=False, encoding='utf-8')

                    #######################
                    #######################
                    
                    # data_list.append({'DEVICE': device, 'FUNCTION': function, 'DATA_TYPE': dataype,
                    #    'NUM_S': result['Scalars'], 'RANGE_S': 0,'NUM_SI': result['Indexes'],'RANGE_SI': 0,
                    #    'NUM_ARR': result['Arrays'],'RANGE_ARR': ranArr, 'NUM_MAT': result['Matrixes'],'RANGE_MAT': ranMat}) 
                        # ignore_index=True)  # 'BOARD': board, 

# df = pd.concat([pd.DataFrame([data]) for data in data_list], ignore_index=True)
```
